```python



'''
i가 1일 때 1-> 3-> [3]
i가 2일 때 2-> 1-> 3-> [3]
i가 3일 때 3-> [3]
i가 4일 때 4-> 7-> 6-> 4 -> [4,7,6]
i가 5일 때 5-> 3-> [3]
'''
def find(parents, x, stack):
    
    ## 자기 자신을 선택하지 않았을 경우
    if parents[x] != x and parents[x] != 0:

        ## 스택안에 선택한 학생이 있을 경우 함수 종료
        if x in stack:
            return stack
        stack.append(x)
        ## 재귀 함수로 자기가 선택한 학생 기준으로 다시 조회
        return find(parents,parents[x], stack)
    
    ## 자기 자신을 선택할 경우, 혹은 선택한 학생이 자기자신을 골랐을 경우
    else:
        return [x]
    

def solution(n,parents):

    answer = 0
    parents.insert(0,0)

    for i in range(1,n+1):
        stack = []
        if parents[i] != 0:
            result = find(parents,i,stack)
        
        ## 결과 리스트에 있는 학생은 완성된 팀으로 처리
        ## 0으로 처리된 학생은 팀에 속하는 것으로 됨
            for i in result:
                parents[i] = 0
            
    
    ## 학생 중 0이 아닌 학생 수 구하기
    for parent in parents:
        if parent != 0:
            answer +=1

    return answer


t = int(input())
for _ in range(t):
    n = int(input())
    parent = list(map(int,input().split()))
    print(solution(n,parent))
```
